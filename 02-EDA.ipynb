```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Clean Dataset
df = pd.read_csv("cleaned_data.csv")
```

```python
# Check Duplicate Values
df.duplicated().sum()
```

```python
#Correlation Matrix
corr = df.corr()

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
#Preparation Time vs Delivery Time
plt.figure(figsize=(7,5))

plt.scatter(df["Preparation_Time_min"], df["Delivery_Time_min"])

plt.xlabel("Preparation Time")
plt.ylabel("Delivery Time")

plt.title("Preparation Time vs Delivery Time")

plt.show()
```

```python
#Courier Experience vs Delivery Time
plt.figure(figsize=(7,5))

plt.scatter(df["Courier_Experience_yrs"], df["Delivery_Time_min"])

plt.xlabel("Courier Experience")

plt.ylabel("Delivery Time")

plt.title("Courier Experience vs Delivery Time")

plt.show()
```

```python
#Weather Count
df["Weather"].value_counts().plot(kind="bar")

plt.title("Weather")

plt.xlabel("Weather Type")

plt.ylabel("Count")

plt.show()
```

```python
#Vehicle Type Count
df["Vehicle_Type"].value_counts().plot(kind="bar")

plt.title("Vehicle Type")

plt.xlabel("Vehicle")

plt.ylabel("Count")

plt.show()
```

```python
#Delivery Time Distribution
plt.figure(figsize=(7,5))
plt.hist(df["Delivery_Time_min"], bins=20)
plt.xlabel("Delivery Time")
plt.ylabel("Frequency")
plt.title("Delivery Time Distribution")
plt.show()
```
